```python
import numpy as np
import matplotlib.pyplot as plt

def divided_diff(x, y):

    n = len(y)
    coef = np.zeros([n, n])
    coef[:, 0] = y

    for j in range(1, n):
        for i in range(n - j):
            coef[i][j] = (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j] - x[i])

    return coef[0, :]

def newton_poly(x_input, x, coef):

    n = len(coef)
    result = coef[0]
    term = 1.0

    for i in range(1, n):
        term *= (x_input - x[i-1])
        result += coef[i] * term

    return result


x = np.array([1, 2, 3])
y = np.array([2, 3, 5])


coef = divided_diff(x, y)


x_new = np.linspace(min(x), max(x), 100)
y_new = [newton_poly(xi, x, coef) for xi in x_new]


print("Koefisien Polinom Newton : \n", coef)

print("\nGambar Grafik : \n\n")

plt.scatter(x, y, color='red', label="Data Asli")
plt.plot(x_new, y_new, color='blue', label="Interpolasi Newton")
plt.title("Interpolasi Polinom Newton 3 titik")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def divided_diff(x, y):
    n = len(y)
    coef = np.zeros([n, n])
    coef[:, 0] = y

    for j in range(1, n):
        for i in range(n - j):
            coef[i, j] = (coef[i+1, j-1] - coef[i, j-1]) / (x[i+j] - x[i])

    return coef[0, :]

def newton_poly(x_input, x, coef):
    n = len(coef)
    result = coef[0]
    term = 1.0

    for i in range(1, n):
        term *= (x_input - x[i-1])
        result += coef[i] * term

    return result

x = np.array([1, 2, 3, 4])
y = np.array([2, 3, 5, 3])

coef = divided_diff(x, y)

x_new = np.linspace(min(x), max(x), 100)
y_new = [newton_poly(xi, x, coef) for xi in x_new]

print("Koefisien Polinom Newton : \n", coef)

print("\nGambar Grafik : \n\n")

plt.scatter(x, y, color='red', label="Data Asli")
plt.plot(x_new, y_new, color='blue', linestyle='-', label="Interpolasi Newton")
plt.title("Interpolasi Polinom Newton 4 titik")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

x = np.array([1, 2, 3])
y = np.array([2, 3, 5])

cs = CubicSpline(x, y)

x_new = np.linspace(min(x), max(x), 100)
y_new = cs(x_new)

print("Koefisien Polinom Kubik : \n", cs.c)
print("\nGambar Grafik : \n\n")

plt.scatter(x, y, color='red', label="Data Asli")
plt.plot(x_new, y_new, color='blue', linestyle='-', label="Interpolasi Cubic Spline")
plt.title("Interpolasi Cubic Spline")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

# **GAME KECIL**




```python
# KARENA GABUT JADI SAYA BIKIN INI, MAIN TEBAK ANGKA

import random

def nebak_angka():
    angka = random.randint(1, 100)
    tebakan = None
    percobaan = 0

    print("Selamat datang di game Tebak Angka!")
    print("Silahkan memilih sebuah angka antara 1 hingga 100. Coba tebak!")

    while tebakan != angka:
        try:
            tebakan = int(input("Masukkan tebakanmu: "))
            percobaan += 1

            if tebakan < angka:
                print("Terlalu kecil! Coba lagi.")
            elif tebakan > angka:
                print("Terlalu besar! Coba lagi.")
            else:
                print(f"Selamat! Kamu berhasil menebak angka {angka} dalam {percobaan} percobaan.")
                break
        except ValueError:
            print("Masukkan angka yang valid!")

if __name__ == "__main__":
    nebak_angka()
```
