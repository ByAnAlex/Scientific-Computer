```python
import numpy as np
import matplotlib.pyplot as plt

# Gaya visual
#plt.style.use('seaborn-poster')

# Definisi parameter
f = lambda t, s: np.exp(-t)
h = 0.1
t = np.arange(0, 1 + h, h)
s0 = -1

# Metode Euler Eksplisit
s = np.zeros(len(t))
s[0] = s0

# Perhitungan Euler
for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h * f(t[i], s[i])

# Plot hasil
plt.figure(figsize=(12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \nfor Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi ODE
f = lambda t, s: np.exp(-t)

# Parameter numerik
h = 0.01
t = np.arange(0, 1 + h, h)
s0 = -1

# Metode Euler Eksplisit
s = np.zeros(len(t))
s[0] = s0

# Iterasi Euler
for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h * f(t[i], s[i])

# Plot hasil
plt.figure(figsize=(12, 8))
plt.plot(t, s, 'b--', marker='o', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \nfor Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Define parameters
f = lambda t, s: t**2 - 4 * s
h = 0.1
t = np.arange(0, 1 + h, h)
s0 = 1

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h * f(t[i], s[i])

# Plotting
plt.figure(figsize=(12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, (1/4) * t**2 - (1/8) * t + (1/32) + (31/32) * np.exp(-4 * t), 'g', label='Exact')
plt.title('Approximate and Exact Solution for ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
